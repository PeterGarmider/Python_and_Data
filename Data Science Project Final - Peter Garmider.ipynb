```python
import pandas as pd
import numpy as np
```

```python
# Source https://www.rug.nl/ggdc/productivity/pwt/ (University of Groningen - Penn Word Table)
t1 = pd.read_excel('pwt90.xlsx', sheetname=2)
t1.head(10)
```

```python
# Select only the years greater than 1980
t1_1980 = t1[t1.year >= 1980]
t1_1980
```

```python
# Select only data for Canada
CAN_t1 = t1_1980[t1_1980.country == 'Canada']
CAN_t1
```

```python
# Select a subset of column data we want to investigate (all rows)
# pop represents total population (millions)
# emp represents number of persons engaged (millions)
# avh represents average hours worked per year
# hc represents human capital index 
CAN_t1_subset = CAN_t1.loc[:,['country', 'year', 'pop', 'emp', 'avh', 'hc']]
CAN_t1_subset
```

```python
# Drop the year column as we already have an index with the same values
CAN_t1_subset = CAN_t1_subset.drop(['year'], axis=1)
```

```python
# I've defined a function for detecting outliers using the interquartile range

def IQR_outliers(series):
    IQR = series.quantile(0.75) - series.quantile(0.25)
    
    lower_bound = series.quantile(0.25) - 1.5*IQR
    print ('Lower Bound: '+ str(lower_bound))
    upper_bound = series.quantile(0.75) + 1.5*IQR
    print ('Upper Bound: '+ str(upper_bound))
    df = series[ (series<lower_bound) | (series>upper_bound) ]
    
    return df
```

```python
# Lets try to fit this data to a line of best fit
from scipy import stats
```

```python
numpy.corrcoef(CAN_t1_subset['emp'], CAN_t1_subset['avh'])
```

```python
m, b, r_value, p_value, slope_std_error = stats.linregress(CAN_t1_subset['emp'], CAN_t1_subset['avh'])
```

```python
plt.scatter(CAN_t1_subset['emp'], CAN_t1_subset['avh']);
plt.plot(CAN_t1_subset['emp'],b + m * CAN_t1_subset['emp'], color='red');
```

```python
# The trend line doesn't seem to represent the data as well as we had hoped
# We can try to use a polynomial fit and see if that results in a better fit
pf0, pf1, pf2, pf3, pf4 = np.polynomial.polynomial.polyfit(CAN_t1_subset['emp'], CAN_t1_subset['avh'], 4)
scatter(CAN_t1_subset['emp'], CAN_t1_subset['avh'])
plt.plot(CAN_t1_subset['emp'],pf0 + \
         pf1*CAN_t1_subset['emp'] + \
         pf2*(CAN_t1_subset['emp']**2) + \
         pf3*(CAN_t1_subset['emp']**3) + \
         pf4*(CAN_t1_subset['emp']**4), color='red');
```

```python
# This follows the data better, but may not neccessarily give a better fit
# Instead lets consider a training, test, and validation set. We will use the K-Folds Cross Valdidation method.
# We do this to measure predictive performance of the model. R^2 - the coefficient of determination is used.
from sklearn import model_selection, metrics
```

```python
kf = model_selection.KFold(n_splits=4,shuffle=True)

set1,set2,set3,set4 = kf.split(CAN_t1_subset)
```

```python
set1
```

```python
c_train1 = CAN_t1_subset.iloc[set1[0],:]
c_test1 = CAN_t1_subset.iloc[set1[1],:]

c_train2 = CAN_t1_subset.iloc[set2[0],:]
c_test2 = CAN_t1_subset.iloc[set2[1],:]

c_train3 = CAN_t1_subset.iloc[set3[0],:]
c_test3 = CAN_t1_subset.iloc[set3[1],:]

c_train4 = CAN_t1_subset.iloc[set4[0],:]
c_test4 = CAN_t1_subset.iloc[set4[1],:]
```

```python
# For a linear test/train/validate:
mc1, bc1, r_c1, p_c1, error_c1 = stats.linregress(c_train1['emp'], c_train1['avh'])
c_train_lin_1 = bc1 + mc1 * c_train1['emp']

mc2, bc2, r_c2, p_c2, error_c1 = stats.linregress(c_train2['emp'], c_train2['avh'])
c_train_lin_2 = bc2 + mc2 * c_train2['emp']

mc3, bc3, r_c3, p_c3, error_c3 = stats.linregress(c_train3['emp'], c_train3['avh'])
c_train_lin_3 = bc3 + mc3 * c_train3['emp']

mc4, bc4, r_c4, p_c4, error_c4 = stats.linregress(c_train4['emp'], c_train4['avh'])
c_train_lin_4 = bc4 + mc4 * c_train4['emp']
```

```python
fig = plt.figure()
gridspec.GridSpec(2,2)

plt.subplot2grid((2,2), (0,0))
scatter(c_train1['emp'], c_train1['avh'],label='Training Data')
scatter(c_test1['emp'], c_test1['avh'], color='red',label='Test Data')
plt.plot(c_train1['emp'],c_train_lin_1, color='purple',label='Regression Model')
plt.legend();

plt.subplot2grid((2,2), (0,1))
scatter(c_train2['emp'], c_train2['avh'],label='Training Data')
scatter(c_test2['emp'], c_test2['avh'], color='red',label='Test Data')
plt.plot(c_train2['emp'],c_train_lin_2, color='purple',label='Regression Model')
plt.legend();

plt.subplot2grid((2,2), (1,0))
scatter(c_train3['emp'], c_train3['avh'],label='Training Data')
scatter(c_test3['emp'], c_test3['avh'], color='red',label='Test Data')
plt.plot(c_train3['emp'],c_train_lin_3, color='purple',label='Regression Model')
plt.legend();

plt.subplot2grid((2,2), (1,1))
scatter(c_train4['emp'], c_train4['avh'],label='Training Data')
scatter(c_test4['emp'], c_test4['avh'], color='red',label='Test Data')
plt.plot(c_train4['emp'],c_train_lin_4, color='purple',label='Regression Model')
plt.legend();
```

```python
# Check the R^2 value for each train/test k-fold split
# R^2 for training set 1
metrics.r2_score(c_train1['avh'], c_train_lin_1)
```

```python
#Absolute error for training set 1
metrics.mean_absolute_error(c_train1['avh'], bc1 + mc1 * c_train1['emp'])
```

```python
# R^2 for test set 1
metrics.r2_score(c_test1['avh'], bc1 + mc1 * c_test1['emp'])
```

```python
# Absolute error for test set 1
metrics.mean_absolute_error(c_test1['avh'], bc1 + mc1 * c_test1['emp'])
```

The R^2 was closer to 1 on the training set than the test set, the Absolute error increased, which we do not want
